```python
sales_order_shared = customers[['Sales_Order', 'Customer_ID']].groupby('Sales_Order')['Customer_ID'].count().rename('Customer_count').to_frame().reset_index()
sales_order_shared
```

```python
unique_order_nrs = sales_order_shared[sales_order_shared['Customer_count'] == 1]['Sales_Order']
shared_order_nrs = sales_order_shared[sales_order_shared['Customer_count'] == 2]['Sales_Order']
display(unique_order_nrs)
display(shared_order_nrs)
```

## Handle unique order nrs

```python
unique_customers = customers[customers['Sales_Order'].isin(unique_order_nrs)]
unique_customers
```

```python
unique_sales_orders = sales_orders[sales_orders['Sales_Order'].isin(unique_order_nrs)]
unique_sales_orders
```

```python
unique_result = unique_customers.join(unique_sales_orders.set_index('Sales_Order'), on='Sales_Order', lsuffix='_customers', rsuffix='_sales_orders')
unique_result
```

## Handle shared order nrs

```python
shared_customers = customers[customers['Sales_Order'].isin(shared_order_nrs)]
shared_customers
```

```python
shared_sales_orders = sales_orders[sales_orders['Sales_Order'].isin(shared_order_nrs)]
shared_sales_orders
```

```python
join_on_columns = ['Sales_Order', 'Item_Position']
shared_result = shared_customers.join(shared_sales_orders.set_index(join_on_columns), on=join_on_columns, how='inner')
shared_result['Item_Position_customers'] = shared_result['Item_Position']
shared_result['Item_Position_sales_orders'] = shared_result['Item_Position']
shared_result = shared_result.drop('Item_Position', axis=1)
shared_result
```

## Handle unmatched customers and sales_orders

```python
pre_full_result = pd.concat([unique_result, shared_result])
pre_full_result
```

### identify unmatched customers

```python
remaining_customers = customers.copy()
remaining_customers['Item_Position_customers'] = remaining_customers['Item_Position']
remaining_customers = remaining_customers.drop('Item_Position', axis=1)
join_on_columns = remaining_customers.columns.to_list()
remaining_customers = remaining_customers.merge(pre_full_result[join_on_columns].set_index(join_on_columns), on=join_on_columns, how='left', indicator=True)
remaining_customers = remaining_customers[remaining_customers['_merge'] == 'left_only'].drop('_merge', axis=1)
remaining_customers
```

### identify unmatche sales_orders

```python
remaining_sales_orders = sales_orders.copy()
remaining_sales_orders['Item_Position_sales_orders'] = remaining_sales_orders['Item_Position']
remaining_sales_orders = remaining_sales_orders.drop('Item_Position', axis=1)
join_on_columns = remaining_sales_orders.columns.to_list()
remaining_sales_orders = remaining_sales_orders.merge(pre_full_result[join_on_columns].set_index(join_on_columns), on=join_on_columns, how='left', indicator=True)
remaining_sales_orders = remaining_sales_orders[remaining_sales_orders['_merge'] == 'left_only'].drop('_merge', axis=1)
remaining_sales_orders
```

### reset Item_Position to 0

```python
remaining_customers = remaining_customers.rename({'Item_Position_customers': 'Item_Position'}, axis=1)
remaining_customers['Item_Position'] = 0

remaining_sales_orders = remaining_sales_orders.rename({'Item_Position_sales_orders': 'Item_Position'}, axis=1)
remaining_sales_orders['Item_Position'] = 0
display(remaining_customers)
display(remaining_sales_orders)
```

### re-match tables

```python
join_on_columns = ['Sales_Order', 'Item_Position']
re_match_result = remaining_customers.join(remaining_sales_orders.set_index(join_on_columns), on=join_on_columns)
re_match_result
```

# Create model

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

import random
random.seed(10)

X = train_data.drop(columns='Reseller')
y = train_data[['Reseller', 'Customer_ID']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
# score for every individual customer
unique_customers = np.unique(X_train['Customer_ID'])

scores = {}
scores_idx = []
for customer in unique_customers:
    c = y_train[y_train['Customer_ID'] == customer]    
    scores[customer] = sum(c['Reseller']) / len(c)
y_train = y_train['Reseller']
y_test = y_test['Reseller']
```

```python
def f(row):  
    if row['Customer_ID'] in scores:
        return scores[row['Customer_ID']]
    return 0
X_train['reseller_score'] = X_train.apply(f, axis=1)
X_test['reseller_score'] = X_test.apply(f, axis=1)
```

```python
# map scores to index
#index_scores = [scores[] for x in X_train_tmp]


c = ['Customer_ID']
X_train_tmp = X_train.drop(columns=c)
X_test_tmp = X_test.drop(columns=c)

pca = PCA(n_components=2)
transformed = pca.fit_transform(X_train_tmp)
print(pca.explained_variance_ratio_)

#transformed
#plt.plot(transformed)
plt.xlabel('Principal Component - 1')
plt.ylabel('Principal Component - 2')
plt.scatter(transformed[:,0],transformed[:,1],c=y_train, marker='.')
plt.show()

import statsmodels.api as sm
from sklearn.metrics import (confusion_matrix,accuracy_score, balanced_accuracy_score)

log_reg = sm.Logit(y_train, transformed).fit()
print(log_reg.summary())

yhat = log_reg.predict(pca.transform(X_test_tmp))
prediction = list(map(round, yhat))

 
# confusion matrix
cm = confusion_matrix(y_test, prediction)
print ("Confusion Matrix : \n", cm)
 
# accuracy score of the model
print('Test accuracy = ', accuracy_score(y_test, prediction))
print('Test balanced accuracy = ', balanced_accuracy_score(y_test, prediction))
```
